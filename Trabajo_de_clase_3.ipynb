#**Muestra**

**2.1. Definicion.**
<br>
Es el subconjunto representativo de la poblacion, seleccionado para analizar y obtener conclusiones generales.

<br>

Ejemplo: 50 Servidores seleccionados para medir latencia en lugar de evaluar 5000 servidores.

<br>

**2.2. Caracteristicas de una buena muestra:**
<br>
Representatividad.
Tamaño demostrativo (n)
Diversidad en los datos.

<br>

**2.3. Importancia en Ingenieria:**
<br>
Reduce costos, tiempo, y esfuerzo, manteniendo un alto nivel de presicion en los resultados.

#**Parametro y Valor Estadistico**

**3.1. Parametro**
<br>
Valor numerico que describe una caracteristica de toda una poblacion. Es el valor rea que nos gustaria conocer, pero que muchas veces es imposible o impractico calcular porque requeriria analizar a toda la poblacion:

<br>

-- La media de calificaciones de TODOS los estudiantes de Ingenieria en Mexico
<br>
-- El porcentaje REAL de computadoras que salen defectuosas de una fabrica
<br>
-- La altura PROMEDIO de TODOS los mexicanos adultos

<br>

**3.2. Valor Estadistico**
<BR>
Es un valor calculado a partir de una muestra (un subconjunto) de la poblacion. Lo usamos para estimar el parametro cuando no podemos medir a toda la poblacion:
<br>
-- La media de calificaciones calculada con una muestra de 100 estudiantes
<br>
-- El porcentaje de defecto encontrado en una mustra de 500 computadoras
<br>
-- La altura promedio medida en una muestra de 1000 mexicanos

<br>

**3.3. Utilidad en la Programacion**
<br>
En programacion esto es similar al momento de realizar pruebas de rendimiento:
<br>
-- Parametro: El tiempo real que tarda tu algoritmo con TODOS los casos posibles
<br>
-- Estadistico: El tiempo medido con una muestra de casos de pruebas

#**Tipos de muestreo**

**4.1. Muestreo Probabilistico (aleatorio)**
<br>
-- Cada elemento tiene igual probabilidad de ser seleccionado.
<br>
-- Tipos:
Aleatorio sistematico: Elegir cada 10°, 15°, etc. Ejemplo: Verificar cada decimo servidor en una cola.
<br>
Aleatorio simple: Numero al azar, herramientas de programacion son utilies para este metodo.
<br>
Estratificado: Dividir la poblacion en subgrupos homogeneos (estratos), seleccionados proporcionalmente.
<br>
Por conglomerados: Dividir en grupos, seleccionar algunos grupos completos.
-- Ventaja: Reduccion de sesgos, resultados mas confiables.

<br>

**4.2. Muestreo no Probabilistico**
-- Seleccion segun criterios no aleatorios.
<br>
-- Tipos:
Conveniencia: Seleccion mas accesible (analisis de servidores locales)
Juicio: Basado en el criterio de un experto
<br>
Cuotas: Seleccion segun proporciones especificas.
<br>
Bola de nieve: Utilizado en poblaciones dificiles de acceder (Usuarios clandestinos de una red)
<br>
-- Riesgo: Mayor posibilidad de riesgos

#**Importacion de Librerias y Dataset**

```python
import pandas as pd

data = pd.read_csv('SoftwareProject_DataSet.csv')
```

```python
data.head()
```

#**Identificar Poblaciones y Muestras**

---



#**Obtener el tamaño de la poblacion**

---



```python
population_size = len(data)
print("Tamaño de la población:", population_size)
```

#**Creacion de una muestra aleatoria simple de 50 registros**

---



```python
sample_size = 50
sample = data.sample(n=sample_size, random_state=42)
```

#**Verificando la muestra**

```python
display(sample)
```

#**Descripcion estadistica de la poblacion completa**

---



```python
population_desc = data[['Costo (USD)', 'Tiempo de desarrollo (horas)']].describe()
print("Descripción de la población completa:")
display(population_desc)
```

#**Muestreo Aletorio sistematico**

---



#**Seleccion sistematica de una muestra**

---



#**Definicion del intervalo K (tamaño poblacional / muestra deseada)**

---



```python
sample_size = 50
interval = population_size // sample_size
```

#**Seleccion de muestra sistematica**

---



```python
systematic_sample = data.iloc[::interval]
```

#**Verificar la Muestra**

```python
print("Muestra obtenida por muestreo sistematico:")
display(systematic_sample)
```

#**Analisis de Subpoblaciones**

---



#**Dividir la poblacion por tipo de proyecto**

---



```python
grouped_by_project_type = data.groupby('Tipo de proyecto')
```

#**Calcular el promedio de lineas de codigo y bugs para cada tipo**

---



```python
group_analysis = grouped_by_project_type[['Líneas de código', 'Número de bugs']].mean()
print("Análisis por tipo de proyecto:")
display(group_analysis)
```

#**Crear una muestra estratificada por 'Tipo de proyecto'**

---



```python
stratified_sample = grouped_by_project_type.sample(n=15, random_state=42)

print("Muestra estratificada por tipo de proyecto:")
display(stratified_sample)
```

```python
data_description = data[['Costo (USD)', 'Tiempo de desarrollo (horas)']].describe()
print("Descripción de la población completa:")
display(data_description)
```

#**Comparacion de Metodos de Muestreo**

---



#**Muestreo por conglomerados basados en "Fase del proyecto"**

---



```python
grouped_by_phase = data.groupby('Fase del proyecto')
conglomerate_sample = grouped_by_phase.sample(n=10, random_state=42)

print("Muestra por conglomerados por fase del proyecto:")
display(conglomerate_sample)
```

#**Muestreo de conveniencia: Seleccion de las primeras 50 filas**

---



```python
convenience_sample = data.head(50)

print("Muestra de conveniencia:")
display(convenience_sample)
```

#**Comparacion de medidas de tendencia central para 'Tiempo de desarrollo (horas)'**

---



```python
methods_comparison = pd.DataFrame({
    'Aleatorio Simple': sample['Tiempo de desarrollo (horas)'].mean(),
    'Conglomerados': conglomerate_sample['Tiempo de desarrollo (horas)'].mean(),
    'Conveniencia': convenience_sample['Tiempo de desarrollo (horas)'].mean(),
    'Poblacion': data['Tiempo de desarrollo (horas)'].mean()
}, index=['Media'])

print("Comparación de medidas de tendencia central:")
display(methods_comparison)
```
